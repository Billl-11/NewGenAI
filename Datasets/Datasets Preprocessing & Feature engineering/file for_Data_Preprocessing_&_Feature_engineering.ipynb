# Features engineering

1. iterate every row
2. check symptoms count for particular row:
    - if symptons count <0 N <= 4 , generate 1 random list
    - if symptons count <4 N <= 8 , generate 2 random list
    - if symptons count <8 N <= 15 , generate 4 random list
    - if symptons count <15 N , generate 5 random list with
4. all generated list will have 4, 5, 6, or 7 symptoms (randomized)
5. all generated list make new dataframe with column called `symptons_list` and also have corresponding label



```python
def get_random_symptons(input_list,number):
    # Shuffle the input list to randomize its order
    randomized_list = random.sample(input_list, len(input_list))

    # Take the first `number` elements from the randomized list
    random_symptons = randomized_list[:number]
    return random_symptons
```

```python
def preprocess(data):
  new_dataframe = []
  for i,row in data.iterrows():
    label = row.loc['Disease']
    count_none = row.isin(['None']).sum()
    count_sympton = 17 - count_none

    list_of_sympton = list(row.iloc[1:(count_sympton+1)])

    # delete underscore
    list_of_sympton = [word.replace('_', ' ') for word in list_of_sympton]

    number = random.choice([4, 5, 6,7])
    list_to_add = get_random_symptons(list_of_sympton,number=number)
    list_to_add = ', '.join(list_to_add)

    if count_sympton > 0 and count_sympton <= 4:
      new_dataframe.append([label, list_to_add])
    elif count_sympton > 4 and count_sympton <= 8:
      for j in range (2):
        new_dataframe.append([label, list_to_add])
    elif count_sympton > 8 and count_sympton <= 15:
      for j in range (4):
        new_dataframe.append([label, list_to_add])
    elif count_sympton > 15:
      for j in range (5):
        new_dataframe.append([label, list_to_add])

  return new_dataframe

```

```python
columns = ['diseases', 'symptons']
new_dataframe = preprocess(df)

df = pd.DataFrame(new_dataframe, columns=columns)
len(df)
```
